## 2 - Multiple Variable Regression

```python
def build_pipeline(numerical_features, categorical_features=None):
    transformers = [("num", "passthrough", numerical_features)]
    if categorical_features:
        transformers.append(("cat", OneHotEncoder(drop="first", sparse_output=False), categorical_features))
    ct = ColumnTransformer(transformers=transformers)
    return Pipeline([("prep", ct), ("model", LinearRegression())])

def eval_pipeline(pipe, X_train_v2, X_test_v3, y_train, y_test, name):
    pipe.fit(X_train, y_train)
    r2_score_train = r2_score(y_train, pipe.predict(X_train))
    r2_score_test = r2_score(y_test, pipe.predict(X_test))
    mae = mean_absolute_error(y_test, pipe.predict(X_test))
    rmse = np.sqrt(mean_squared_error(y_test, pipe.predict(X_test)))
    print(f"{name:<45}  Train R2 = {r2_score_train:.4f}  Test R2={r2_score_test:.4f} MAE={mae:.1f}")
    return pipe, r2_score_train, r2_score_test, mae, rmse

print("Model comparison - Multiple Regression subsets")
print("="*90)
model_results = {}

configs = [
    ("Model 1: WSP",                      ["WSP"],                          None),
    ("Model 2: WSP + AGE",                ["WSP","AGE"],                    None),
    ("Model 3: WSP + AGE + HID",          ["WSP","AGE","HID"],              None),
    ("Model 4: WSP + AGE + SPC",          ["WSP","AGE"],                    ["SPC"]),
    ("Model 5: WSP + AGE + HID + SPC",    ["WSP","AGE","HID"],              ["SPC"]),
    ("Model 6: All features",             ["WSP","MASS","AGE","SPD","HID"], ["SPC"]),
]

for name, num_feat, cat_feat in configs:
    pipe = build_pipeline(num_feat, cat_feat)
    p, r2tr, r2te, mae, rmse = eval_pipeline(pipe, X_train, X_test, y_train, y_test, name)
    model_results[name] = dict(pipe=p, r2tr=r2tr, r2te=r2te, mae=mae, rmse=rmse)
```

Multiple regression models were explored using various combinations of available variables 

* 50.80% and 50.00% data explainability was achieved when using Wingspan (WSP) as a predictor on both the training and testing set respectively.

* An addition of Age to the previous predictor (WSP+AGE) yielded a very marginal increase to both scores in the training and testing sets.

* Introducing hide thickness (HID) to the existing model (WSP+AGE+HID) kind of increased the training score with an addition of about 0.0001% but did the opposite in the testing set of which that caused an increase in the MAE in comparison to the results obtained previously.

* Substituting HID with species(SPC) (WSP + AGE + SPC) increased the training score but the testing score decremented with MAE further increasing again. The introduction of this variable poetentially brings about noise instead of additional signal, this suggest that there is a very low correlation relation between SPC and FHO when the addition of AGE and WSP are in play.

* Including HID as well altogether with SPC ( WSP + AGE + HID + SPC) increased the scores in both the training and testing set.

* When utilising all the features, the final score for the training set was at 52.02% and on the testing had a very huge decrement with an r2 score standing at 43.87%. This resulted to a huge jump in the mean absolute error to 238.3.

The best performing model is M2 which is the utilisation of WSP and AGE as predictors only, these two features produced an r2 score of 49.53% on the testing set and 50.82% on the training set of which this led to the small difference between the two being 1.29 meaning that the generalisation of the model is better in comparison to the other modeling results. \ 

The results obtained in model 6 show signs of overfitting with the lowest test r2 score of 0.4387m the inroduction of SPD and MASS clearly provided a negative impact in the performance of the model. The difference between the test score and train score is the highest in comparison to other results obtained.
